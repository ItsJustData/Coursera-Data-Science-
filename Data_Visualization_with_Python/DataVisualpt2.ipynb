```python
df_topsrv = pd.read_csv('https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/coursera/Topic_Survey_Assignment.csv')
df_topsrv
```

```python
df_topsrv.rename(columns={'Unnamed: 0' : ''}, inplace=True)
df_topsrv.set_index('',inplace=True)
df_topsrv
```

```python
df_topsrv.sort_values(['Very interested'], ascending=False, axis=0, inplace=True)
df_topsrv
```

```python
df_topsrv.dtypes
```

```python
df_topsrv = df_topsrv.astype(float)
```

```python
df_topsrv
```

```python
df_topsrv2 = (df_topsrv / 2233)
```

```python
df_topsrv3 = round(df_topsrv2*100, 2)
```

```python
df_topsrv3.columns = list(map(str, df_topsrv3.columns))
df_topsrv3
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

ax = df_topsrv3.plot(kind='bar', figsize=(20, 8), width = 0.8, color = ['#5cb85c','#5bc0de','#d9534f'], fontsize = 14)
ax.set_title('Percentage of Respondents\' Interest in Data Science Areas', fontsize = 16)
ax.legend(fontsize = 14)
ax.axes.get_yaxis().set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate('{:,.2f}%'.format(height), (p.get_x()+.55*width, p.get_y() + height + 2), ha = 'center',fontsize=14) 


```
